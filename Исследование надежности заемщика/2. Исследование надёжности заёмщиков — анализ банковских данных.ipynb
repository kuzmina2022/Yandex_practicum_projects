## Предобработка данных

###  Заполнение пропусков

Выше мы выяснили, что имеются пропуски в данных с общим трудовым стажем и ежемесячным доходом, и они составляют 10%.

Посмотрим на строки с пропусками, чтобы понять их причину.

```python
data_nan = data[data['days_employed'].isnull() & data['total_income'].isnull()]
display(data_nan)
```

```python
#посмотрим тип занятости в пропущенных значениях
data_nan['income_type'].unique() 
```

Среди пропусков нет безработных клиентов, вероятно, пропуски связаны с тем, что клиенты получают заработную плату либо пенсию в данном банке и поэтому в анкетах не указали доход.

Заменим пропущенные значения. В случае с количественными переменными наилучшие решение для замены - метод медианы `median()`. 
Среднее значение некорректно характеризует данные, когда некоторые значения сильно выделяются среди большинства (особенно, когда рассматриваются заработные платы), поэтому метод среднего арифметического нам не подойдет. 

```python
#посчитаем медианные значения для трудового стажа и дохода
days_employed_median = data['days_employed'].median()
print(days_employed_median)
total_income_median = data['total_income'].median()
print(total_income_median)
```

```python
#заменяем пропуски на полученные медианные значения
data['days_employed'].fillna(days_employed_median, inplace=True)
data['total_income'].fillna(total_income_median, inplace=True)
```

```python
data.head(15)
```

**Выводы**

В данных имеются пропуски в колонках с трудовым стажем и доходом, их количество совпадает. Возможно, причина пропусков в том, что клиенты, получающие пенсию либо зарплату в данном банке, не указали в анкете свой доход. Пропуски мы заменили на медианные значения.

###  Проверка данных на аномалии и исправления.

В данных встречаются артефакты (аномалии) — значения, которые не отражают действительность и появились по какой-то ошибке. Например, отрицательное количество дней трудового стажа или слишком большие значения в столбце `days_employed`. Вероятно, это связано с техническим сбоем при загрузке данных.

Уберем отрицательные значения в столбце со стажем, применим к нему функцию `abs()`.

```python
data['days_employed'] = abs(data['days_employed'])
```

Проверим, есть ли значения больше 22000. Трудовой стаж больше 22000 дней (это примерно 60 лет) вряд ли возможен, клиенты старше 75 лет не могут обращаться за кредитом.

```python
data.loc[(data['days_employed'] > 22000)]
```

Значение стажа не слишком важно для исследования, аномальные значения заменим медианой. 

```python
data.loc[(data['days_employed'] > 22000), 'days_employed'] = data['days_employed'].median() 
```

```python
# Проверим, что замена прошла успешно
data.loc[(data['days_employed'] > 22000)]
```

Проверим возраст заемщиков, он не должен быть меньше 18 или больше 75 лет. 

```python
data.loc[(data['dob_years'] < 18) | (data['dob_years'] > 75)] 
```

```python
data.loc[(data['dob_years'] < 18) | (data['dob_years'] > 75), 'dob_years'].unique() 
```

Мы видим, что в колонке `dob_years` есть аномальные значения, все они равны 0. Вероятно, возраст просто не указан или утерян. Этот показатель не очень важен для исследования, заменим его медианным значением.

```python
data.loc[(data['dob_years'] < 18) | (data['dob_years'] > 75), 'dob_years'] = data['dob_years'].median()
```

```python
# Проверяем, что аномалий по возрасту нет
data.loc[(data['dob_years'] < 18) | (data['dob_years'] > 75), 'dob_years'].unique() 
```

Проверим, есть ли необычные значения среди дохода. Посмотрим значения ниже прожиточного минимума или слишком большие. 

```python
data.loc[(data['total_income'] < 12000) | (data['total_income'] > 10_000_000)] 
```

В `total_income` артефактов не обнаружено. 

Теперь посмотрим значения в `children`, количество детей не может быть меньше 0, также проверим большие значения.

```python
data.loc[(data['children'] < 0) | (data['children'] > 9)]
```

```python
data.loc[(data['children'] < 0) | (data['children'] > 9), 'children'].unique()
```

Мы видим здесь аномальные значения -1 и 20. Вероятно, это техничский сбой. 
Для исследования нам важно знать количество детей у заемщика, поэтому эти данные мы не можем заменить каким-то средним значением. Но т.к. этих данных всего 123 на 21000 (<1%), их удаление не повлияет значительно на результат исследования. 

Удаляем строки с неподходящими значениями и проверяем результат.

```python
data = data.drop(data[(data['children'] < 0) | (data['children'] > 9)].index).reset_index(drop=True)
```

```python
data.loc[(data['children'] < 0) | (data['children'] > 9), 'children'].unique()
```

```python
data['gender'].unique()
```

```python
data.loc[(data['gender'] == 'XNA')]
```

Есть одно странное значение пола, удалим его.

```python
data = data.drop(data[data['gender'] == 'XNA'].index).reset_index(drop=True)
```

```python
data.loc[(data['gender'] == 'XNA')]
```

**Выводы**

Мы выполнили проверку данных на артефакты. Отрицательные и слишком большие значения трудового стажа, а также ошибки в возрасте заменили на медианные значения. Вероятной причиной возникновения такиих ошибок могут быть технические сбои при обработке или передаче данных.

Аномальные данные также обнаружены в колонке с количеством детей заемщика. Эти данные важны для исследования, поэтому мы не могли их заменить, но т.к. число артефактов не превышает 1%, мы удалили их без значительного ущерба для релузьтатолв исследования. ###  Категоризация дохода.

У нас большое количество разных вариантов дохода. 
Работать с единичными отрывками и делать из них статистические выводы нельзя. Значит, нужно сгруппировать данные, чтобы численности каждой группы хватало для формулировки выводов.
Нужна категоризация — объединение данных в категории. 

Создадим столбец `total_income_category` с категоризацией доходов:
     
* 0–30000 — `'E'`;
* 30001–50000 — `'D'`;
* 50001–200000 — `'C'`;
* 200001–1000000 — `'B'`;
* 1000001 и выше — `'A'`

```python
# Для начала напишем функцию, которая разделит значения доходов по категориям
def income_category(income):
    if  income <= 30000:
        return 'E'
    if income <= 50000:
        return 'D'
    if income <= 200_000:
        return 'C'
    if income <= 1000_000:
        return 'B'
    return 'A'
  
print(income_category(200000))
print(income_category(1000001))
```

```python
# Теперь создадим новый столбец, применяя к ['total_income'] функцию income_category
data['total_income_category'] = data['total_income'].apply(income_category)
data.head(10)
```

```python
data['total_income_category'].value_counts()
```

Наиболее распространенная категория доходов среди клиентов банка  - `C` от 50 до 200 тыс. 

###  Категоризация целей кредита.

Теперь объединим различные цели получения кредита по категориям.
Создадим функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:
* `'операции с автомобилем'`,
* `'операции с недвижимостью'`,
* `'проведение свадьбы'`,
* `'получение образования`'

```python
# Посмотрим, какие есть цели кредита в наших исходных данных
data['purpose'].sort_values().unique()
```

```python
def purpose_category(purpose):
    if 'авто' in purpose:
        return 'операции с автомобилем'
    if 'жиль' in purpose or 'недвиж' in purpose:
        return 'операции с недвижимостью'
    if 'свадьб' in purpose:
        return 'проведение свадьбы'
    if 'образов' in purpose: 
        return 'получение образования'
    return 'цель неопределена'
    
    
print(purpose_category('сыграть свадьбу'))    
print(purpose_category('на покупку подержанного автомобиля'))
print(purpose_category('на покупку жилья'))
print(purpose_category('строительство недвижимости'))
print(purpose_category('сыграть свадьбу'))
print(purpose_category('получение дополнительного образования'))
print(purpose_category('купить лотерейные билеты'))
    
```

```python
data['purpose_category'] = data['purpose'].apply(purpose_category)
data.head()
```

```python
data['purpose_category'].value_counts()
```

Среди целей кредита лидируют операции с недвижимостью.
