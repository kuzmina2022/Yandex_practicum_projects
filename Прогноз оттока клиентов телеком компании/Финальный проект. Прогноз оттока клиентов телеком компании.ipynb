# Телеком — задача проекта

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

### Описание услуг

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

### Описание данных

Данные состоят из файлов, полученных из разных источников:

- `contract.csv` — информация о договоре;
- `personal.csv` — персональные данные клиента;
- `internet.csv` — информация об интернет-услугах;
- `phone.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

## Подготовка данных для обучения моделей

Делим признаки на обучающую и тестовую выборки в соотношении 3:1. Валидационную выборку не используем, будем применять кросс-валидацию.

```python
features_train, features_test, target_train, target_test = train_test_split(
    data.drop('left', axis=1), data.left, test_size=0.25, random_state=120922) 
print('Размеры обучающей выборки:', features_train.shape, target_train.shape)
print('Размеры тестовой выборки:', features_test.shape, target_test.shape)
```

Кодируем категориальные признаки методом OHE.

```python
# Список категориальных признаков для кодирования
data.select_dtypes(include='object').columns
```

```python
f_train_ohe = pd.get_dummies(features_train, drop_first=True)
```

```python
f_train_ohe
```

```python
f_test_ohe = pd.get_dummies(features_test, drop_first=True)
```

```python
f_test_ohe
```

Масштабируем числовые данные, используя StandartScaler.

```python
numeric = ['MonthlyCharges','TotalCharges','Duration']
scaler = StandardScaler()
scaler.fit(f_train_ohe[numeric])
f_train_ohe[numeric] = scaler.transform(f_train_ohe[numeric])
f_test_ohe[numeric] = scaler.transform(f_test_ohe[numeric])
```

```python
f_train_ohe
```

```python
f_test_ohe
```

Рассмотрим распределение целевого признака.

```python
plt.figure(figsize=(5, 5))
sns.countplot(y="left", data=data)
plt.title('Распределение целевого признака', fontsize=20);
```

```python
data.left.mean()
```

Соотношение покинувших компанию клиентов к действующим примерно 1:3. Для учета дибаланса при обучении моделей будем использовать гиперпараметр class_weight='balanced'.

Данные разделены на обучающую и тестовую выборки в отношении 3:1, категориальные признаки закодированы, числовые  масштабированы, дисбаланс учтен.

## Тестирование моделей

```python
#Создаем таблицу для записи результатов моделей
result_df = pd.DataFrame(columns=['Parameters', 'AUC-ROC'])
result_df
```

```python
def result_write(model_name, params, roc_auc):
    """Функция записывает результаты предсказания в таблицу"""
    result_df.loc[model_name,'Parameters'] = params
    result_df.loc[model_name,'AUC-ROC'] = roc_auc
    return result_df
```

### Логистическая регрессия

```python
logreg = LogisticRegression(random_state=RANDOM_STATE, class_weight='balanced')
roc_auc= cross_val_score(logreg, f_train_ohe, target_train, scoring='roc_auc', cv=10).mean().round(4)
accuracy = cross_val_score(logreg, f_train_ohe, target_train, scoring='accuracy', cv=10).mean().round(4)

print("Logistic Regression")
print("AUC-ROC =", roc_auc)
print('Accuracy = ', accuracy)
```

```python
logreg = LogisticRegression(random_state=RANDOM_STATE)
roc_auc_2= cross_val_score(logreg, f_train_ohe, target_train, scoring='roc_auc', cv=10).mean().round(4)
accuracy_2 = cross_val_score(logreg, f_train_ohe, target_train, scoring='accuracy', cv=10).mean().round(4)

print("Logistic Regression")
print("AUC-ROC =", roc_auc)
print('Accuracy = ', accuracy)
```
