### Описание проекта

Получена задача проанализировать данные для интернет-магазина, который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation) в период до 2016 года. Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.


**Задачи проекта:**

* Ознакомиться с данными
* Провести предобработку данных и посчитать суммарные продажи во всех регионах;
* Провести исследовательский анализ с помощью графиков;
* Составить портет пользователя в каждом регионе продажи игр;

**Проверка гипотез:**

* Проверить гипотезу о том, средние пользовательские рейтинги платформ Xbox One и PC одинаковы;
* Проверить гипотезу о том, что средние пользовательские рейтинги жанров Action и Sports разные;
* Подвести итоги проделанной работы.

*Цель проекта - выявить закономерности, определяющие успешность игры. Это позволит спланировать рекламную кампанию и сделать ставку на потенциально популярный продукт.*

В целом, в данных нет каких-то необычных значений, кроме значения `tbd` в оценках пользователей. 

**Что означают данные рейтинга.**

Рейтинг игр ESRB: 

**«EC»** («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет.

**«E»** («Everyone») — «Для всех»: Содержание вполне подходит для возрастной категории от 6 лет. Первоначально **"K-A"** ("Kids to Adults")


**«E10+»** («Everyone 10 and older») — «Для всех от 10 лет и старше»


**«T»** («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет. 

**«M»** («Mature») — «Для взрослых»: Материалы игры не подходят для подростков младше 17 лет.

**«AO»** («Adults Only 18+») — «Только для взрослых».

**«RP»** («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.

**TBD** - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено). Используется, если какая-то информация еще не определена или решение по вопросу не принято.

```python
data_grafic
```

```python
# Построим график распределения продаж по годам
plt.figure(figsize=(20, 10))
sns.lineplot( data=data_grafic, x="year_of_release", y="total_sales", hue='platform', ci=None
                     ).set_title('График продаж')

plt.xlabel('Год')
plt.ylabel('Сумма продаж')
plt.xticks(np.arange(1980 ,2016, step=2));
```

По графику можно предположить, что наиболее характерный срок существования игровой платформы 9-11 лет.

### Зависимость продаж от оценок критиков и пользователей

Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. 

Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами. 

Рассмотрим платформу Xone. 

```python
sns.scatterplot(data=data_actual.query('platform == "xone"'), x='critic_score', y='total_sales', 
                color='purple').set(title='Зависимость продаж от оценки критиков для Xone');
```

```python
sns.scatterplot(data=data_actual.query('platform == "xone"'), x='user_score', 
                y='total_sales', color='indigo').set(title='Зависимость продаж от оценки пользователей для Xone');
```

По рассеиванию заметно, что оценки критиков больше влиют на продажи, чем оценки пользователей.

Вычислим корреляцию между отзывами и продажами.

```python
# формируем датасеты, чтобы вычислитьпо ним корреляцию
data_critic = data_actual.query('platform == "xone"')[['critic_score', 'total_sales']].copy()
data_users = data_actual.query('platform == "xone"')[['user_score', 'total_sales']].copy()
```

```python
# корреляция продаж и оценок критиков
data_critic['critic_score'].corr(data_critic['total_sales'])
```

```python
# корреляция продаж и оценок пользователей
data_users['user_score'].corr(data_users['total_sales'])
```

Вычисленная корреляция подтверждает наш вывод по диаграммам рассеивания: влияние оценки критиков на продажи умеренные, оценка пользователей не влияет на продажи.

Проверим наши выводы на других платформах.

```python
# Используем цикл, чтобы пройтись по всем выбранным потенциально прибыльным платформам
platform_list = ['wiiu', 'ps4', '3ds', 'psv', 'pc']

# В цикле строим диаграммы рассеивания продаж от оценок критиков и пользователей, а так же считаем корреляции
for platform in platform_list: 
    data_plot=data_actual.query('platform == @platform')
           
    fig, ax = plt.subplots(1, 2, figsize=(15,5))    
    
    sns.scatterplot(data=data_plot, x='critic_score', y='total_sales', color='purple', ax=ax[0]
                   ).set(title='Зависимость продаж от оценок критиков')
    sns.scatterplot(data=data_plot, x='user_score', y='total_sales', color='indigo', ax=ax[1]
                   ).set(title='Зависимость продаж от оценок пользователей')
    fig.suptitle(platform)    
    plt.show()
    
    crit_corr = data_plot['critic_score'].corr(data_plot['total_sales'])
    
    if  0 <= abs(crit_corr) <=0.1:
        print('Корреляция общего числа продаж и оценки критиков несущественная и равна:', crit_corr)
    if  0.1 < abs(crit_corr) <=0.3:
        print('Корреляция общего числа продаж и оценки критиков слабая и равна:', crit_corr)
    if  0.3 < abs(crit_corr) <=0.6:
        print('Корреляция общего числа продаж и оценки критиков средняя и равна:', crit_corr)
    if  0.6 < abs(crit_corr) <=0.9:
        print('Корреляция общего числа продаж и оценки критиков высокая и равна:', crit_corr)
    if  0.9 < abs(crit_corr) <=1:
        print('Корреляция общего числа продаж и оценки критиков очень высокая и равна:', crit_corr)
        
    
    user_corr = data_plot['user_score'].corr(data_plot['total_sales'])
    
                                                 
    if  0 <= abs(user_corr) <= 0.1:
        print('Корреляция общего числа продаж и оценки пользователей несущественная и равна:', user_corr)
    if  0.1 < abs(user_corr) <= 0.3:
        print('Корреляция общего числа продаж и оценки пользователей слабая и равна:', user_corr)
    if  0.3 < abs(user_corr) <=0.6:
        print('Корреляция общего числа продаж и оценки пользователей средняя и равна:', user_corr)
    if  0.6 < abs(user_corr) <=0.9:
        print('Корреляция общего числа продаж и оценки пользователей высокая и равна:', user_corr)
    if  0.9 < abs(user_corr) <=1:
        print('Корреляция общего числа продаж и оценки пользователей очень высокая и равна:', user_corr)
    print('\n')
                                             
```

По платформам зависимости в целом похожи:

* WiiU - средняя зависимость продаж от оценок критиков и пользователей
* PS4 - зависимость продаж от оценок критиков средняя, оценки пользователей не влияют на продажи
* 3DS - зависимость продаж от оценок критиков средняя, от оценки пользователей - слабая
* PSV - продажи не зависят от оценок
* PC - слабая зависиомсть от оценок


#### Вывод

В целом, можно сделать вывод, что оценки критиков умеренно влияют на продажи игр, а оценки пользователей - слабо.

## Проверка гипотез

Необходимо проверить следующие гипотезы:

* Средние пользовательские рейтинги платформ **Xbox One** и **PC** одинаковые;

* Средние пользовательские рейтинги жанров **Action** (англ. «действие», экшен-игры) и **Sports** (англ. «спортивные соревнования») разные.



### Гипотеза:  Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Проверяем гипотезу, что средние пользовательские рейтинги платформ **Xbox One** и **PC** одинаковые.

```python
# Выбираем данные, необходимые для проведения теста
data_xone = data_actual.query('platform == "xone"')['user_score']
data_pc = data_actual.query('platform == "pc"')['user_score']
```

```python
data_xone
```

```python
# В данных есть пропуски, которые нужны удалить для корректности дальнейших расчетов
data_xone.dropna(inplace=True)
```

```python
data_xone
```

```python
data_pc
```

```python
data_pc.dropna(inplace=True)
```

```python
data_pc
```

Посчитаем среднее, дисперсию и стандартное отклонение среди пользовательских рейтингов для платформ Xbox One и PC.

```python
print('Средний пользовательский рейтинг для платформы Xbox One: {0:1.2f}'.format(data_xone.mean()))
print('Стандартное отклонение равно {0:1.3f}'.format(data_xone.std()))
print('Дисперсия: {0:1.3f}'.format(np.var(data_xone)))
```

```python
print('Средний пользовательский рейтинг для платформы PC: {0:1.2f}'.format(data_pc.mean()))
print('Стандартное отклонение равно {0:1.3f}'.format(data_pc.std()))
print('Дисперсия: {0:1.3f}'.format(np.var(data_pc)))
```

**Нулевая гипотеза H₀** : "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые".

**Альтернативная гипотеза H₁** : "Средние пользовательские рейтинги платформ Xbox One и PC отличаются".

**Пороговое значение alpha** 5%

Нулевая гипотеза всегда формимулируется так, чтобы использовать знака равенства, в нашем случае это равенство средних пользовательских рейтингов. 
Исходя из H₀ формулируется альтернативная гипотеза H₁. Это утверждение, которое принимается верным, если отбрасывается H₀, т.е. в альтернативной гипотезе предполагаем, что средние пользовательские рейтинги не равны. 
Критерием для проверки гипотез является сравнение p-value (вероятности получить наблюдаемый или еще более далекий от предполагаемого результат при условии, что нулевая гипотеза верна) и выбранным уровнем значимости. 

В Python есть специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: `scipy.stats.ttest_ind (array1, array2, equal_var)`, его мы и используем.
Параметр `equal_var` принимаем `False`, т.к. у нас разные размеры выборок и значения дисперсии.

```python
alpha = 0.05

results = st.ttest_ind(
    data_xone,
    data_pc,
    equal_var=False
)   

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Нулевая теория не отвергнута.
Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно 55%. Это явно слишком большая вероятность, чтобы делать вывод о значимом различии между средними значениями рейтинга.

### Гипотеза: Средние пользовательские рейтинги жанров Action и Sports разные.

Проверяем гипотезу, что средние пользовательские рейтинги жанров **Action** и **Sports** разные. 

```python
# Выберем данные, необходимые для анализа
data_action = data_actual.query('genre == "action"')['user_score']
data_sports = data_actual.query('genre == "sports"')['user_score']
```

```python
print('Средний пользовательский рейтинг для жанра Action: {0:1.2f}'.format(data_action.mean()))
print('Стандартное отклонение равно {0:1.3f}'.format(data_action.std()))
print('Дисперсия: {0:1.3f}'.format(np.var(data_action)))
```

```python
print('Средний пользовательский рейтинг для жанра Sports: {0:1.2f}'.format(data_sports.mean()))
print('Стандартное отклонение равно {0:1.3f}'.format(data_sports.std()))
print('Дисперсия: {0:1.3f}'.format(np.var(data_sports)))
```

**Нулевая гипотеза H₀** : "Средние пользовательские рейтинги жанров **Action** и **Sports** равны".

**Альтернативная гипотеза H₁** : "Средние пользовательские рейтинги жанров **Action** и **Sports** отличаются.".

**Пороговое значение alpha** 5%

Параметр `equal_var` принимаем `False`, т.к. у нас разные размеры выборок и значения дисперсии.

```python
alpha = 0.05

results = st.ttest_ind(
    data_action,
    data_sports, 
    equal_var=False)   

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Нулевая гипотеза отвергнута, значит, вероятность получить близкие по значению средние величины рейтинга для жанров **Action** и **Sports** крайне мала.
